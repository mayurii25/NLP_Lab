```python
#TOKENIZATION

import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize

text = "I believe this would help the reader to understand how tokenization     works. as well as realize its importance."

sents = sent_tokenize(text)
print(sents)

print(word_tokenize(text))

words = [word_tokenize(sent) for sent in sents]
print(words)
```

```python
#STOPWORDS

import nltk
nltk.download('stopwords')
nltk.download('punkt')

from nltk.tokenize import TreebankWordTokenizer
from nltk.corpus import stopwords
from string import punctuation

text = " I believe this would help the reader understand how tokenization works. as well as realize its importance (text)."

custom_list = set(stopwords.words("english") + list(punctuation))

tokenizer = TreebankWordTokenizer()
word_list = [word for word in tokenizer.tokenize(text) if word not in custom_list]
print(word_list)
```

```python
#N-GRAM

import nltk
nltk.download('stopwords')
nltk.download('punkt')

from nltk.collocations import BigramCollocationFinder, TrigramCollocationFinder

word_list = ['I', 'believe', 'this', 'would', 'help', 'the', 'reader', 'to', 'understand', \
              'how', 'tokenization', 'works', 'as', 'well', 'as', 'realize', 'its', 'importance']
finder = BigramCollocationFinder.from_words(word_list)
print(finder.ngram_fd.items())
find = TrigramCollocationFinder.from_words(word_list)
print(find.ngram_fd.items())
```

```python
import nltk
nltk.download('punkt')
from nltk.util import ngrams

word_list = ['I', 'believe', 'this', 'would', 'help', 'the', 'reader', 'to', 'understand', \
              'how', 'tokenization', 'works', 'as', 'well', 'as', 'realize', 'its', 'importance']

n=3

for gram in ngrams(word_list, n):
  print(gram)
```

```python
#STEMMING

from nltk.tokenize import word_tokenize
from nltk.stem.lancaster import LancasterStemmer

new_text = "It is important to by very pythonly while you are pythoning \
    with python. All pythoners have pythoned poorly at least once."
l_s = LancasterStemmer()
stem_lan = [l_s.stem(word) for word in word_tokenize(new_text)]
print(stem_lan)
```

```python
## Word sense disambiguation

import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet   # used when we want to find contextual meaning of a word
for ss in wordnet.synsets('mouse'):
    print(ss, ss.definition())

from nltk.wsd import lesk
from nltk.tokenize import word_tokenize

context_1 = lesk(word_tokenize("Sing in a lower tone, along with the bass"), "bass")
print(context_1, context_1.definition())
context_2 = lesk(word_tokenize("The sea bass is really very hard to catch"), "bass")
print(context_2, context_2.definition())
context_3 = lesk(word_tokenize("My mouse is not working, need to change it"),"mouse")
print(context_3, context_3.definition())
```

```python
# CountVectorizer

import pandas as pd
corpus = [
    'This is the first document from heaven',
    'but the second document is from mars',
    'And this is the third one from nowhere',
    'Is this the first document from nowhere'
]
df = pd.DataFrame({'Text': corpus})
print(df)

from sklearn.feature_extraction.text import CountVectorizer
count_v = CountVectorizer()
X = count_v.fit_transform(df['Text']).toarray()
print(count_v.get_feature_names_out())
print(X)
print(count_v.vocabulary_)

count_v2 = CountVectorizer(stop_words=['this', 'is'])
Y = count_v2.fit_transform(df['Text']).toarray()
print(count_v2.get_feature_names_out())
print(Y)
```

```python
# TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer
corpus = [
    'This is the first document from heaven',
    'but the second document is from mars',
    'And this is the third one from nowhere',
    'Is this the first document from nowhere'
]

vectorizer = TfidfVectorizer()
vectorizer.fit(corpus)
print(vectorizer.vocabulary_)
print(vectorizer.idf_)
```

```python
# Hashing Vectorizer

import pandas as pd
from sklearn.feature_extraction.text import HashingVectorizer
corpus = [
    'This is the first document from heaven',
    'but the second document is from mars',
    'And this is the third one from nowhere',
    'Is this the first document from nowhere'
]
df = pd.DataFrame({'Text': corpus})
hash_v = HashingVectorizer(n_features=9, norm=None, alternate_sign=False)
# norm is set to None to avoid normalization, and alternate_sign is set to False to avoid negative values
X_hash = hash_v.fit_transform(df['Text']).toarray()
print(X_hash)
```
